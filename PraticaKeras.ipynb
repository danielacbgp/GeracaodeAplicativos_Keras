##1.2 Transformando os campos do tipo *Object*

###1.2.1 Transformando o campo *Species* em One Hot Encoded 

```
from sklearn.preprocessing import LabelEncoder
# encode class values as integers
from keras.utils import np_utils
encoder = LabelEncoder()
encoder.fit(iris.Species)
encoded_Species = encoder.transform(iris.Species)
# convert integers to dummy variables (i.e. one hot encoded)
dummy_Species = np_utils.to_categorical(encoded_Species)
```

###1.2.2 Retirando o campo *Specie* 

```
base_final = iris.drop(['Species'], axis=1)
```

#2 Dividindo os Dados em Treinamento (train) e Teste (test) 

Podemos dividir nosso dataframe (**base_final**) em dois:

- Uma parte para treinar nosso modelo (80% da base): **train**
- Uma parte para fazer previsões e testar nosso modelo (20% da base): **test**
